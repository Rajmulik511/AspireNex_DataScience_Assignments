```python
x= creditdata.drop('Class',axis=1)
y= creditdata['Class']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,f1_score,precision_score,recall_score,confusion_matrix
```

Dictionary a way to store multiple machine learning classifiers in a single data structure. This can simplify the process of training, testing, and comparing multiple models within a loop or function.

```python
classifier ={
    'LogisticRegression':LogisticRegression(),
    # 'RandomForestClassifier':RandomForestClassifier(),
    'DecisionTreeClassifier':DecisionTreeClassifier()
}
for name,clf in classifier.items():
  print('------------',name,'model -----------')
  clf.fit(x_train,y_train)
  y_pred=clf.predict(x_test)
  print('Accuracy',accuracy_score(y_test,y_pred))
  print('Precision',precision_score(y_test,y_pred))
  print('Recall',recall_score(y_test,y_pred))
  print('F1_score',f1_score(y_test,y_pred))
```

Undersampling


```python
normal= creditdata[creditdata['Class']==0]
fraud= creditdata[creditdata['Class']==1]
normal.shape,fraud.shape
```

```python
normal_sample=normal.sample(n=473) #randomly Select 473 rows form Normal
normal_sample.shape
```

```python
classifier ={
    'LogisticRegression':LogisticRegression(),
    # 'RandomForestClassifier':RandomForestClassifier(),
    'DecisionTreeClassifier':DecisionTreeClassifier()
}
for name,clf in classifier.items():
  print('------------',name,'model -----------')
  clf.fit(x_train,y_train)
  y_pred=clf.predict(x_test)
  print('Accuracy',accuracy_score(y_test,y_pred))
  print('Precision',precision_score(y_test,y_pred))
  print('Recall',recall_score(y_test,y_pred))
  print('F1_score',f1_score(y_test,y_pred))
```

Oversampling

```python
x=creditdata.drop('Class',axis=1)
y=creditdata['Class']
x.shape,y.shape
```

```python
from imblearn.over_sampling import SMOTE
```

```python
x_resample,y_resample=SMOTE().fit_resample(x,y)
```

```python
y_resample.value_counts()
```

stsrt from 37.35 in video

```python
x_train, x_test, y_train, y_test = train_test_split(x_resample, y_resample, test_size=0.2, random_state=42)
```

```python
classifier ={
    'LogisticRegression':LogisticRegression(),
    # 'RandomForestClassifier':RandomForestClassifier(),
    'DecisionTreeClassifier':DecisionTreeClassifier()
}
for name,clf in classifier.items():
  print('------------',name,'model -----------')
  clf.fit(x_train,y_train)
  y_pred=clf.predict(x_test)
  print('Accuracy',accuracy_score(y_test,y_pred))
  print('Precision',precision_score(y_test,y_pred))
  print('Recall',recall_score(y_test,y_pred))
  print('F1_score',f1_score(y_test,y_pred))
```
